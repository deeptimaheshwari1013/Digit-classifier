```python
# importing required libraries + MNIST database

import tensorflow as tf
import matplotlib.pyplot as plt

(x_train, y_train) , (x_test,y_test) = tf.keras.datasets.mnist.load_data()
```

```python
# preprocess the data (normalizing the input data)

x_train = x_train / 255.0
x_test = x_test/255.0
```

```python
# defining the nn model

model = tf.keras.models.Sequential([
    tf.keras.layers.Flatten(input_shape = (28,28)), # 2D image pixels -> 1D vector
    tf.keras.layers.Dense(128, activation = 'relu'),
    tf.keras.layers.Dense(10, activation = 'softmax') # output layer for 10 digis
])
```

```python
# adam adjusts the learning rate automatically

model.compile(optimizer = 'adam',
              loss = 'sparse_categorical_crossentropy',
              metrics = ['accuracy'])
```

```python
model.fit(x_train, y_train, epochs = 5)

test_loss, test_acc = model.evaluate(x_test, y_test)
print(f"Test accuracy: {test_acc}")
```
